## Data Preprocessing

```python
import os
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.transform import resize
```

```python
x = []
y = []

DIRECTORY = "Images/"
CATEGORIES = ['cow', 'sheep']

for category in CATEGORIES:
    path = os.path.join(DIRECTORY, category)
    for img in os.listdir(path):
        img = imread(os.path.join(path,img))
        img = resize(img, (160,160,3))
        x.append(img)
        y.append(CATEGORIES.index(category))

x = np.array(x)
y = np.array(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
combined_data = list(zip(x_train, y_train))
np.random.shuffle(combined_data)

shuffled_features, shuffled_labels = zip(*combined_data)
x_train = np.array(shuffled_features)
y_train = np.array(shuffled_labels)
```

## Loading the Pretrained Model

```python
import tensorflow as tf
from tensorflow import keras
import tensorflow.keras.layers as tfl

base_model = tf.keras.applications.densenet.DenseNet121(
    include_top=False,
    weights='imagenet',
    input_shape=(160,160,3),
)

base_model.trainable = False

model = keras.models.Sequential()

model.add(tf.keras.Input((160,160,3)))

model.add(base_model)

model.add(tfl.GlobalAveragePooling2D())

model.add(tfl.Dense(1, activation = 'sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate = 0.001),
              loss = 'binary_crossentropy',
              metrics = ['accuracy'])
```

```python
model.fit(x_train, y_train, epochs = 5, batch_size = 5, validation_data = (x_test, y_test))
```

```python
model.save('saved_model')
```

```python
test_img = imread("test.jpg")
test_img = resize(test_img, (160,160,3))
test_img = np.expand_dims(test_img, axis=0)
loaded_model = tf.keras.models.load_model('saved_model')
predictions = loaded_model(test_img)
```

```python
predicted_class_labels = tf.argmax(predictions, axis=1)
```
